### Handling missing values

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=["int64","float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns
```

### Error / Outlier Handling

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) |
          (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

### Hypothesis Testing

```python
df[["temp", "cnt"]].dtypes
```

```python
df["temp"] = pd.to_numeric(df["temp"], errors="coerce")
df["cnt"]  = pd.to_numeric(df["cnt"], errors="coerce")
```

```python
df = df.dropna(subset=["temp", "cnt"])
```

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df["temp"], df["cnt"])
corr, p_value
```

### Encoding (Categorical â†’ Numerical) One-Hot Encoding

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

X = df_encoded.drop("cnt", axis=1)
y = df_encoded["cnt"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
import numpy as np
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2, mse, rmse, mae
```

### Polynomial Regression

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Example numeric features in bike_rental dataset
num_features = ["temp", "atemp", "hum", "windspeed"]

# Select numeric features
X_num = df[num_features].copy()
y = df["cnt"].copy()
```

```python
# Convert any invalid strings to numeric (coerce errors to NaN)
for col in X_num.columns:
    X_num[col] = pd.to_numeric(X_num[col], errors="coerce")

# Drop rows with NaN in X_num or y
X_num = X_num.dropna()
y_clean = y.loc[X_num.index]  # Keep target aligned
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_poly, y_clean, test_size=0.2, random_state=42
)
```

```python
lr_poly = LinearRegression()
lr_poly.fit(X_train, y_train)

y_pred = lr_poly.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print(f"R2 Score: {r2:.4f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```

### Lasso & Ridge Regression (Best Alpha)

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV

alphas = {"alpha":[0.001,0.01,0.1,1,10,100]}

ridge = GridSearchCV(Ridge(), alphas, cv=5)
ridge.fit(X_train, y_train)

lasso = GridSearchCV(Lasso(max_iter=5000), alphas, cv=5)
lasso.fit(X_train, y_train)

ridge.best_params_, lasso.best_params_
```

```python
best_lasso = lasso.best_estimator_
y_pred_lasso = best_lasso.predict(X_test)

r2_score(y_test, y_pred_lasso),
mean_squared_error(y_test, y_pred_lasso),
np.sqrt(mean_squared_error(y_test, y_pred_lasso)),
mean_absolute_error(y_test, y_pred_lasso)
```
