### Glucose

Normal: 70–99 mg/dL

Prediabetes: 100–125 mg/dL

Diabetes: ≥ 126 mg/dL

Most important indicator for early diabetes detection

### Blood Pressure

Typical Range: 50 – 120

Most healthy individuals fall within 60–80 mmHg.
Higher blood pressure values (>80) may indicate hypertension, which can worsen insulin resistance over time.

Very low values are rare and do not show a strong direct link with diabetes.

Blood pressure alone is not a strong predictor of diabetes, but it supports overall metabolic risk assessment.


### Blood Pressure

Typical Range: 10 – 60

Lower values (<20) indicate lower body fat levels.

Values between 20–35 represent a healthier fat distribution.

Higher values (>35) suggest increased subcutaneous fat, often associated with insulin resistance.

Higher skin thickness values are more commonly observed in diabetic patients, though the feature has moderate predictive power.


### Insulin

Typical Range: 15 – 850

Lower insulin values generally indicate normal insulin function.

Moderate values (100–300) suggest insulin resistance.

Extremely high values (>300) often appear in diabetic individuals, though variability is high.

Insulin shows high variability but provides useful information after missing value treatment and outlier handling.

### BMI

Typical Range: 18 – 50

Normal BMI (18.5–24.9) is associated with lower diabetes risk.

Overweight (25–29.9) individuals show increased risk.

Obese individuals (BMI ≥ 30) have a significantly higher likelihood of diabetes.

BMI is one of the most important lifestyle-related predictors of diabetes.

### Diabetes Pedigree Function

Typical Range: 0.05 – 2.5

Lower values indicate minimal genetic influence.

Moderate values reflect some family history of diabetes.

Higher values (>0.7) suggest strong genetic susceptibility.

Genetic factors can increase diabetes risk even when other clinical features appear normal.


### Age

Typical Range: 21 – 80

Younger individuals (<30) show relatively low diabetes prevalence.

Risk increases between 30–45 years and rises sharply after 45 years.

Older individuals show higher diabetes incidence.

Age amplifies the effect of glucose and BMI on diabetes risk.

### Outcome (Target Variable)

0 → Non-Diabetic

1 → Diabetic

Individuals with higher glucose, BMI, age, and genetic risk are far more likely to be classified as diabetic.








## 3. missing values anaalysis and tratment

```python
df.isnull().sum()
```

```python
import numpy as np

cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)
```

```python
df[cols_with_zero].isnull().sum()
```

```python
# Glucose
df['Glucose'] = df['Glucose'].fillna(df['Glucose'].median())

# BloodPressure
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].median())

# SkinThickness
df['SkinThickness'] = df['SkinThickness'].fillna(df['SkinThickness'].median())

# Insulin
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].median())

# BMI
df['BMI'] = df['BMI'].fillna(df['BMI'].median())
```

```python
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].isnull().sum()
```

## 4. outlier analysis and treatment

```python
def remove_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]

df = remove_outliers_iqr(df)
```

```python
Q1 = df["Glucose"].quantile(0.25)
Q3 = df["Glucose"].quantile(0.75)
```

```python
IQR = Q3 - Q1

# Step 3: Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

```

```python
glucose_outliers = df[
    (df["Glucose"] < lower_bound) |
    (df["Glucose"] > upper_bound)
]
print("Number of Outliers:", glucose_outliers.shape[0])
```

```python
from sklearn.preprocessing import MinMaxScaler
cols_to_scale = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Outcome', axis=1)
y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred_train=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_train,y_pred_train)
```
